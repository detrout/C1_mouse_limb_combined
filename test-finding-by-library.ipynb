```python
import generate_combined_transcript_C1
import to_include
from woldrnaseq.models import load_library_tables
import os
from woldrnaseq.make_tracks import make_bigwig_track_name, make_bam_track_name
import pandas
import find_bigwigs
```

```python
clusters = pandas.DataFrame(find_bigwigs.read_peng_20180710_cluster_memberships())
```

```python
libraries = [os.path.expanduser(x.strip()) for x in generate_combined_transcript_C1.ASOF_RUN17_library_files.split('\n')]
```

```python
library_df = load_library_tables(libraries)
library_df = library_df.reindex(clusters['cell_id'])
library_df.index.name = 'library_id'
print(library_df.shape)
```

```python
for library_id, row in library_df.iterrows():
    track_name = make_bam_track_name(row, row.analysis_dir)
    track_path = os.path.join(row.analysis_dir, track_name)
    track_exists = os.path.exists(track_path)
    index_exists = os.path.exists(track_path+'.bai')
    #print(library_id, track_path, track_exists, index_exists)
          
```
