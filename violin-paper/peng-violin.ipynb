```python
from argparse import ArgumentParser
import pandas
import seaborn
import matplotlib
from matplotlib import pylab
```

```python
%matplotlib inline
```

```python
gene_index = 'gene_index'
cell_type = 'cell_type'
```

```python
filename = 'peng-example-violin.txt'
data = pandas.read_csv(filename, header=0, index_col=None)
print(data.head())
```

```python
fig = pylab.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1)
seaborn.violinplot(x=cell_type, y=gene_index, colors='colors', data=data, scale='count', ax=ax)
fig.savefig('peng-example-violinplot.svg')
```

```python
pengs_colors = [
    (.5,0,0),
    (1,0,0),
    (0,1,0),
    (0,0,0),
    (.8,.8,.8),
    (0,1,1),
    (1,.5,0),
    (.9,.9,0),
    (.5,.2,.6),
    (0,0,1),
    (.5,.6,.2),
    (1,0,1),
]
pengs_cluster_names = [
    'Muscle 3',
    'Muscle 2',
    'Muscle 1',
    'Mesenchyme 2 + unknown',
    'Mesenchyme 1',
    'Chondrocyte',
    'EMP',
    'Macrophage',
    'Endothelial',
    'Perichondrium',
    'Epithelial',
    'Neural crest',
]
p = seaborn.color_palette(palette=pengs_colors)
```

```python
with p:
    fig = pylab.figure(figsize=(14,10))
    ax = fig.add_subplot(1, 1, 1)
    seaborn.violinplot(x=cell_type, y=gene_index, cut=0, bw=.2, data=data, linewidth=.2, scale='count', ax=ax)
    ax.set_ylabel('')
    ax.set_xlabel('')
    ax.set_xticklabels(pengs_cluster_names, rotation=45)
    #fig.savefig('peng-example-violinplot.svg')
    #fig.savefig('peng-example-violinplot.png')
    #fig.savefig('peng-example-violinplot.eps')
```
