```python
gene_coverage_by_library = {}
for c in models.load_all_gene_coverage(libraries):
    gene_coverage_by_library[c.name] = c
```

```python
experiment_genes_considered = {}
experiment_libraries_by_genes = {}
genes_considered = []
for experiment_name, experiment in experiments.iterrows():
    for r in experiment['replicates']:
        genes_considered.append(gene_coverage_by_library[r].shape[0])
    experiment_genes_considered[experiment_name] = genes_considered
    experiment_libraries_by_genes[experiment_name] = (
        len(genes_considered), 
        numpy.median(genes_considered),
        numpy.std(genes_considered)
    )
```

```python
experiment_libraries_by_genes = pandas.DataFrame(experiment_libraries_by_genes).T
experiment_libraries_by_genes.columns = ['N', 'mean genes', 'st dev genes']
experiment_libraries_by_genes
```

```python
def make_coverage_plot(experiment, coverage):
    with pyplot.style.context('seaborn-dark-palette'):
        f = pyplot.figure(dpi=100)
        ax = f.add_subplot(1, 1, 1)
        coverage.plot(ax=ax)
        ax.set_title('Coverage for {}'.format(experiment))
        ax.set_xlabel("position quantile (5' to 3')")
        ax.set_ylabel('Read depth')
        ylim_bottom, ylim_top = ax.get_ylim()
        ax.set_ylim(0, ylim_top)
        ax.legend(bbox_to_anchor=(1.05, 1),
                  loc=2,
                  borderaxespad=0.0)
    return f

def make_median_normalized_summary(ax, experiment, library_ids, coverage, fillcolors=None, bare=False, fontname='Ariel', fontsize=16):
    with pyplot.style.context('seaborn-dark-palette'):
        centered = coverage[library_ids] / coverage[library_ids].median(axis=0)

        fillcolor = fillcolors[experiment] if fillcolors is not None else 'blue'
        if fillcolor in ('black', '#000000'):
            linecolor='#e0e0e0'
        else:
            linecolor='black'
        median = centered.median(axis=1)
        stdev = centered.std(axis=1)
        s = ax.plot(median, color=linecolor, linewidth=1)
        
        ax.text(50, 0.25, s=experiment,
                horizontalalignment='center',
                fontname=fontname,
                fontsize=fontsize
        )
        
        ax.fill_between(
            x=median.index,
            y1=median-stdev,
            y2=median+stdev,
            label='+/- std. dev',
            facecolor=fillcolor,
            edgecolor='black',
            alpha=0.75
            )
```
