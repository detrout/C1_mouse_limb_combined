```python
tsv_software = []
tsv_derived_from = []

for name, row in metadata.iterrows():
    if row['file'].startswith('EN'):
        obj = server.get_json(row['file'])
        tsv_software.append([(x['software']['name'], x.get('version')) for x in obj['analysis_step_version']['software_versions']])
        tsv_derived_from.append(obj['derived_from'])
```

```python
bam_software = []
bam_derived_from = []

for accessions in tsv_derived_from:
    obj = server.get_json(accessions[-1])
    bam_software.append([(x['software']['name'], x.get('version')) for x in obj['analysis_step_version']['software_versions']])
    bam_derived_from.append(obj['derived_from'])
```
