```python
import pandas
import os
import sys
```

```python
HTSW = os.path.expanduser('~diane/proj/htsworkflow')
if HTSW not in sys.path:
    sys.path.append(HTSW)
from htsworkflow.submission import encoded
```

```python
server = encoded.ENCODED('test.encodedcc.org')
server.load_netrc()
```

```python
book = pandas.ExcelFile('processed_files_test-uploaded2.xlsx')
files = book.parse('File')
```

```python
for i, row in files.iterrows():
    obj = server.get_json(row.accession)
    if obj['dataset'] != row.dataset:
        print(i, 'replacing', obj['dataset'], 'with', row.dataset)
        server.patch_json(obj['@id'], {'dataset': row.dataset})
        
```

```python
for i, row in files.iterrows():
    obj = server.get_json(row.accession)
    if obj['assembly'] != row.assembly:
        print(i, 'replacing', obj['assembly'], 'with', row.assembly)
        server.patch_json(obj['@id'], {'assembly': row.assembly})
        
```
